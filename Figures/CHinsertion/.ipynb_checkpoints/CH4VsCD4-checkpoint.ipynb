```python
import numpy as np
import matplotlib.pyplot as plt

width = 3.33*2
height = width/1.4   # 1.618 is the golden ratio
one_col = (width, height)
fig, ax = plt.subplots(figsize=one_col)

x1,y1 = np.loadtxt('no081d.dat',usecols=[2,3],unpack=True)
x2,y2 = np.loadtxt('no081f.dat',usecols=[2,3],unpack=True)

a = np.amax(y1)
print(a)
y1 -= a

b = np.amax(y2)
y2 -= b

a = np.amin(y1)
y1 /= a

b = np.amin(y2)
y2 /= b


#y2 *= 3
#y1 -= 0.01
#x1 += 0.1
y2 *= -1



ax1 = plt.subplot(2,2,1)
ax1.plot(x1,y1,color='r',label='CH4')
ax1.legend(loc=1,frameon=False,prop={'size': 12})
#ax.xlabel('Mass (u)')
ax1.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the bottom edge are off
    right=False,         # ticks along the top edge are off
    labelleft=False)
ax1.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=True,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False,
    direction='in')
ax1.set_ylim(-0.005,0.3)
ax1.set_xlim(150,175)
fig.set_size_inches(*one_col)
fig.subplots_adjust(left=0.2, bottom=0.19, right=0.96, top=0.96)

ax2 = plt.subplot(2,2,3)
ax2.plot(x2,y2,color='k',label='CH4')
ax2.legend(loc=4,frameon=False,prop={'size': 12})
#ax.xlabel('Mass (u)')
ax2.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the bottom edge are off
    right=False,         # ticks along the top edge are off
    labelleft=False)
ax2.set_ylim(-0.3,0.005)
ax2.set_xlim(150,175)
fig.set_size_inches(*one_col)
fig.subplots_adjust(left=0.2, bottom=0.19, right=0.96, top=0.96)


x3,y3 = np.loadtxt('nt117d_masstrace1.dat',unpack=True)
a,b,x4,y4 = np.loadtxt('np114a_mass.dat',unpack=True)

a = np.amax(y3)
print(a)
y3 -= a

b = np.amax(y4)
y4 -= b

a = np.amin(y3)
y3 /= a

b = np.amin(y4)
y4 /= b

x3 -= 335
#x4 -= 335

x3 /= 361
#x2 /= 361

x3 *= 12
#x4 *= 12
x3 += 154
#x4 += 154

y4 *= -3
#y3 *= 0.5
y3 -= 0.01
x3 += 0.1

ax3 = plt.subplot(2,2,2)
ax3.plot(x3,y3,color='r',label='CD4')
ax3.legend(loc=1,frameon=False,prop={'size': 12})
#ax.xlabel('Mass (u)')
ax3.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the bottom edge are off
    right=False,         # ticks along the top edge are off
    labelleft=False)
ax3.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=True,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False,
    direction='in')
ax3.set_ylim(-0.01,0.6)
ax3.set_xlim(150,175)
fig.set_size_inches(*one_col)
fig.subplots_adjust(left=0.2, bottom=0.19, right=0.96, top=0.96)

ax4 = plt.subplot(2,2,4)
ax4.plot(x4,y4,color='k',label='CD4')
ax4.legend(loc=4,frameon=False,prop={'size': 12})
#ax.xlabel('Mass (u)')
ax4.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the bottom edge are off
    right=False,         # ticks along the top edge are off
    labelleft=False)
ax4.set_ylim(-0.6,0.01)
ax4.set_xlim(150,175)
fig.set_size_inches(*one_col)
fig.subplots_adjust(left=0.2, bottom=0.19, right=0.96, top=0.96, wspace=0.2, hspace=0)

fig.add_subplot(111, frameon=False)
# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.xlabel('Mass (u)',size=12,labelpad=10)
plt.savefig('CH4VsCD4.pdf',dpi=800)

plt.show()
```
