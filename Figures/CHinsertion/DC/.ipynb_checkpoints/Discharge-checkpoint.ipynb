```python
#this is with the 78% ACYN sample
import numpy as np
import matplotlib.pyplot as plt

width = 3.33
height = width/1.4   # 1.618 is the golden ratio
one_col = (width, height)
fig, ax = plt.subplots(figsize=one_col)

a,b,x3,y3 = np.loadtxt('no081d.dat',unpack=True)
x1,y1 = np.loadtxt('no081e.dat',usecols=[2,3],unpack=True)
x2,y2 = np.loadtxt('no081f.dat',usecols=[2,3],unpack=True)

a = np.amax(y1)
print(a)
y1 -= a

b = np.amax(y2)
y2 -= b

c = np.amax(y3)
y3 -= c

a = np.amin(y1)
y1 /= -a
y1 += 0.005

b = np.amin(y2)
y2 /= b
y2 -= 0.01

c = np.amin(y3)
y3 /= c

plt.plot(x2,y2,color='r',label="DC on",linewidth=1)
plt.plot(x1,y1,color='k',label="DC off",linewidth=1)
#plt.plot(x3,y3,color='b')
plt.xlabel('Mass (u)')
plt.ylabel('Relative Intensity (arb.)')
plt.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=True,      # ticks along the bottom edge are off
    right=False,         # ticks along the top edge are off
    labelleft=True)
plt.legend(loc=1,frameon=False)
plt.xlim(140,190)
plt.ylim(-0.3,0.3)
fig.set_size_inches(*one_col)
fig.subplots_adjust(left=0.2, bottom=0.19, right=0.96, top=0.96)
plt.savefig('DC1.pdf',dpi=800)
plt.show()

width = 3.33
height = width/1.4   # 1.618 is the golden ratio
one_col = (width, height)
fig, ax = plt.subplots(figsize=one_col)

y3 -= 0.01
y3 *= -1

plt.plot(x2,y2,color='r',label="DC on",linewidth=1)
plt.plot(x3,y3,color='k',label="1+1 REMPI",linewidth=1)
#plt.plot(x3,y3,color='b')
plt.xlabel('Mass (u)')
plt.ylabel('Relative Intensity (arb.)')
plt.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=True,      # ticks along the bottom edge are off
    right=False,         # ticks along the top edge are off
    labelleft=True)
plt.legend(loc=1,frameon=False)
plt.xlim(140,190)
plt.ylim(-0.3,0.3)
fig.set_size_inches(*one_col)
fig.subplots_adjust(left=0.2, bottom=0.19, right=0.96, top=0.96)
plt.savefig('DC2.pdf',dpi=800)
plt.show()


```
